# matplotlib - plotting library

matplotlib is a plotting library for python. It is quite powerful at the same time it is hard to learn.

```python
import matplotlib.pyplot as plt
```

To get the plots in the notebook, you have to type in the following line.

```python
%matplotlib inline
```

```python
plt.plot([1, 10, 5, 4])
```

You can pass in both x and y values.

```python
import numpy as np
```

```python
x = np.arange(100)
y = x**3 - 80*x**2 + 20*x*2 - 5
plt.plot(x,y)
```

You can control the plot display by showing the labels and legends.

```python
x = np.linspace(1, 2.5, 20)
plt.plot(x, x**2, 'r-', x, x**3, 'go', x, x**0.5, 'bs')
plt.ylabel('my value')
plt.xlabel('some x param')
plt.legend(['square', 'cube', 'root'])
```

```python
x = np.linspace(1, 2.5, 5)
plt.bar(x, x**2, width=0.3)
```
